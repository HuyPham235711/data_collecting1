## Import Libraries

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style='whitegrid')
```

## Load the Processed Data

```python
# Load the processed data
df = pd.read_csv('../data/processed/TSLA1.csv')
df.head()
```

## Data Overview

```python
# Get basic statistics and information
df.info()
df.describe()
```

## Creating a Time Index

```python
# Assuming the index of the DataFrame corresponds to the time series
# Creating a new time index based on the row index
df['Time'] = pd.date_range(start='2024-10-31', periods=len(df), freq='5T')  # 5 minutes frequency
df.set_index('Time', inplace=True)

```

## Plot Closing Prices Over Time

```python
# Plotting closing prices over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x=df.index, y='close', marker='o')
plt.title('TSLA Closing Prices Over Time')
plt.xlabel('Time')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

## Plot Distribution of Closing Prices

```python
# Plotting the distribution of closing prices
plt.figure(figsize=(10, 6))
sns.histplot(df['close'], bins=30, kde=True)
plt.title('Distribution of TSLA Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```
